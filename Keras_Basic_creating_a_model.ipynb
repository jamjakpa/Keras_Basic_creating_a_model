```python
import numpy as np 
```

```python
from sklearn.datasets import load_iris
```

```python
#create variable iris
iris = load_iris()
```

```python
#kind of object return
type(iris)
```

```python
#print out attribute
print(iris.DESCR)
```

```python
# my X feature
X = iris.data
```

```python
#4 measurments lined up for instance. length width sepal and pedal length and width
X
```

```python
#grab the labels
y = iris.target
```

```python
#what we see here is class 0 class 1 and class 2
y
```

```python
#class 0 -->[1,0,0]
#class 1 -->[0,1,0]
#class 2 -->[0,0,1]
```

```python
#transform each of these in vector of 3 
from keras.utils import to_categorical
```

```python
# now i cast it
y = to_categorical(y)
```

```python
#150 instances and now it has 3 values per label
y.shape
```

```python
#now I have setup y
y
```

```python
#now i can split up the data to training set and test set
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.33,random_state=42)
```

```python
#now we have the 4 components X_train ( training data)
#the X_test (testing features)
#y_train (labels for the training data)
#y_test ( labels for the test data )
#everything is now been shuffled and mixed
```

```python
#scale and standarized your data its a good idea
from sklearn.preprocessing import MinMaxScaler
```

```python
#this is just an explaination what MinMaxScaler does is take that list and divide it by the max value wich is 20 
#everything is between zero and 1,because it has been divided by the max value in this case 20
#np.array([5,10,15,20])/20
```

```python
#creating scaler object
scaler_object = MinMaxScaler()
```

```python
#then i fit it, only to the training data
scaler_object.fit(X_train)
```

```python
#scaled version of 
scaled_X_train = scaler_object.transform(X_train)
```

```python
scaled_X_test = scaler_object.transform(X_test)
```

```python
#all the values in our training data are between 0 and 1
#scaled_X_train
```

```python
#now I am building the network with keras
from keras.models import Sequential
from keras.layers import Dense
```

```python
#adding layers to the model (dense layer)
#adding output layer with 3 neurons i choose softmax
#each particular layer has different parameters 8,8,3 are neurons in each layer
model = Sequential() #creating Sequential model
model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dense(8,input_dim=4,activation='relu'))
model.add(Dense(3,activation='softmax')) # [0.2.,0.3,0.5]
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
#asking summary of the model  with mode.summary()
model.summary()
```

```python
#i sclae the mode data  X an dY train 1 epoch is running through the entire dataset
#verbose how much output of information , i run it now with 150 epochs
#result should be around 90%
model.fit(scaled_X_train,y_train,epochs=150,verbose=2)
```
